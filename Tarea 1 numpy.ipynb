## 2. Mostrar las primeras filas del DataFrame para verificar que los datos se han cargado correctamente 

```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


base = pd.read_csv('titanic.csv')

base.head()
```

## 3. Ver la información básica del DataFrame

```python
base.info
```

## 4. Resumen estadístico de las columnas numéricas



## 

```python
base.describe()
```

## 5. Histograma de las edades de los pasajeros

```python
plt.hist(base['Age'], bins='auto', density=True, histtype='step',color='skyblue', edgecolor='black')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.title('Histograma de Edades')
plt.show
```

## 6. Diagrama de dispersión entre la tarifa y la edad

```python
plt.scatter(base['Fare'], base['Age'], color='blue', alpha=0.5)
plt.xlabel('Tarifa')
plt.ylabel('Edad')
plt.title('Diagrama de Dispersión: Tarifa vs. Edad')
plt.show
```

## 7. Gráfico de barras de los pasajeros que sobrevivieron vs los que no

```python
survived_counts = base['Survived'].value_counts()

# Crear el gráfico de barras
survived_counts.plot(kind='bar', color=['blue', 'red'])

# Añadir etiquetas y título
plt.xlabel('Sobrevivió')
plt.ylabel('Cantidad de Pasajeros')
plt.title('Pasajeros que sobrevivieron vs. No sobrevivieron')

# Cambiar las etiquetas del eje x
plt.xticks([0, 1], ['No', 'Sí'], rotation=0)

# Mostrar la gráfica
plt.show()
```

## 8. Encuentra todos los pasajeros femeninos de clase 1.



```python
pasajeros_femeninos_clase_1 = base[(base['Sex'] == 'female') & (base['Pclass'] == 1)]

# Mostrar los resultados
pasajeros_femeninos_clase_1.head()
```

## 9. Encuentra todos los pasajeros menores de 18 años.



```python
pasajeros_menores_de_18 = base[base['Age'] < 18]

pasajeros_menores_de_18.head()
```

## 10. Encuentra todos los pasajeros que sobrevivieron y tienen más de 50 años.

```python
pasajeros_mayores_50 = base[(base['Survived']==1) &(base['Age']>50) ]


pasajeros_mayores_50.head()
```
